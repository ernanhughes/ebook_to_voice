# Clone the repo and install the dependencies

```python
!rm -rf audiblez
!git clone  https://github.com/santinic/audiblez
!apt-get -y install espeak-ng ffmpeg libgtk-3-dev
!cd audiblez && pip install poetry && poetry install
```

# Download an ebook and convert it to audiobook

```python
!wget "https://archive.org/download/AnimalFarmByGeorgeOrwell/Animal%20Farm%20by%20George%20Orwell.epub" -O book.epub
!cd audiblez && poetry run audiblez ../book.epub --cuda
```

# Try the audiobook

```python
from IPython.display import Audio
!cp audiblez/book.m4b book.mp4
display(Audio('book.mp4', autoplay=True))
```
